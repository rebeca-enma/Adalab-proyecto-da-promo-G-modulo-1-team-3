## Getting Started

These instructions will give you a copy of the project up and running on
your local machine for development and testing purposes. See deployment
for notes on deploying the project on a live system.

### Prerequisites

Los juegos están diseñados para Python

### Installing

Los juegos están accesibles en https://github.com/Adalab/promo-g-project-module-1


Instrucciones de Ejecución
Asegúrate de tener Python instalado en tu sistema.
Descarga el archivo del código fuente del proyecto ahorcado.py.
Abre una terminal o línea de comandos.
Navega hasta el directorio donde se encuentra el archivo ahorcado.py.
Ejecuta el programa escribiendo python ahorcado.py y presiona Enter.
Sigue las instrucciones en pantalla para jugar al ahorcado.
