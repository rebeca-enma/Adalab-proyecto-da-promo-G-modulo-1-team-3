```python
import random


class Juegos:

    def imprimir_tablero(tablero):
        print("******************")
        print(tablero[0][0],"|",tablero[0][1],"|",tablero[0][2])
        print("__________")
        print(tablero[1][0],"|",tablero[1][1],"|",tablero[1][2])
        print("__________")
        print(tablero[2][0],"|",tablero[2][1],"|",tablero[2][2])

    def tres_en_raya(self):
        jugar = True
        tablero = [[' ' for _ in range(3)] for _ in range(3)]
        jugada= 0
        ficha_jugador1= "X"
        ficha_jugador2= "O"


        while jugar:
            if jugada %2 == 0:
                ficha_tablero = ficha_jugador1
                
            else:
                ficha_tablero = ficha_jugador2

            try:
                fila = int(input(f"Introduce FILA {ficha_tablero}"))
                columna = int(input(f"Introduce COLUMNA {ficha_tablero}"))

            except:
                print("Dato erroneo, solo puede introducir NUMEROS del 1 al 2")
                continue

            
            if fila <0 and fila >2:
                print("Error, valor de FILA incorrecto (selecciona un numero 0, 1 o 2)")
                continue
            if columna <0 and columna >2:
                print("Error, valor de COLUMNA incorrecto (selecciona un numero 0, 1 o 2)")
                continue

            if tablero[fila][columna] != " ":
                print("Casilla ocupada")
                self.imprimir_tablero(tablero)
                continue # Que no ponga ficha si la casilla ya esta ocupada
                

            tablero[fila][columna]= ficha_tablero

                
            if tablero[0][0] == tablero[0][1] and tablero[0][0] == tablero[0][2] and tablero[0][0] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[1][0] == tablero[1][1] and tablero[1][0] == tablero[1][2] and tablero[1][0] != " ":
                    if tablero[1][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[2][0] == tablero[2][1] and tablero[2][0] == tablero[2][2] and tablero[2][0] != " ":
                    if tablero[2][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False
                    
            elif tablero[0][0] == tablero[1][0] and tablero[0][0] == tablero[2][0] and tablero[0][0] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[0][1] == tablero[1][1] and tablero[0][1] == tablero[2][1] and tablero[0][1] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[0][2] == tablero[1][2] and tablero[0][2] == tablero[2][2] and tablero[0][2] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[0][0] == tablero[1][1] and tablero[0][0] == tablero[2][2] and tablero[0][0] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False

            elif tablero[0][2] == tablero[1][1] and tablero[0][2] == tablero[2][0] and tablero[0][2] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 🎉")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 🏆")
                    jugar = False
            
            elif jugada == 8:        
                print("Vaya, parece que los jugadores han quedado en tablas!😰")
                jugar = False
            

            self.imprimir_tablero(tablero)
            jugada += 1

    def piedra_papel_tijera(self):
         
        lista_opciones = ["Piedra","Papel","Tijera"]

        partidas_maximas = 3
        contador_jugador1 = 0
        contador_jugador2 = 0
        partida = 0

        ganador = max(contador_jugador1,contador_jugador2)


        while partida < partidas_maximas:
            jugador2 = input("Elige para jugar Piedra, Papel, Tijera").lower()
            jugador1 = random.choice(lista_opciones).lower()
        

            if jugador1 == jugador2:
                print("Oh, habeis quedado EMPATE!")
            
            elif (jugador1 == "piedra" and jugador2 == "tijera") or (jugador1 == "papel" and jugador2 == "piedra") or (jugador1 == "tijera" and jugador2 == "papel"):
                
                contador_jugador1 +=1 
                print("¡Partida ganada por el ordenador!")

            else:
                contador_jugador2 +=1
                print("¡Partida ganada por el jugador!")
            
            partida = contador_jugador1 + contador_jugador2


            

            print(f"Partidas jugadas: {partida}")
        print("**********************************")
        if contador_jugador1 > contador_jugador2:
            print("Ha ganador EL ORDENADOR!😭")
        else:
            print("ENHORABUENA!!! Has ganado 🏆")

    
    def ahorcado(self):
        
        palabra_adivinar = input("elige una palabra").lower() #jugador 1
        letras_falladas = []
        vidas= 6
        palabra_incognito = ["_"]*len(palabra_adivinar)
        secreta = ""
        muñeco= ["------\n|  |\n|  O\n| /|\\\n| / \\\n|------\n",
                "------\n|  |\n|  O\n| /|\\\n| / \n|------\n",
                "------\n|  |\n|  O\n| /|\\\n| \n|------\n",
                "------\n|  |\n|  O\n| /|\n| \n|------\n",
                "------\n|  |\n|  O\n| /\n| \n|------\n",
                "------\n|  |\n|  O\n| \n| \n|------\n",
                "------\n|  |\n|  \n| \n| \n|------\n"]


        while vidas > 0 and secreta!= palabra_adivinar:

                letra_jugada = input("Introducce una letra").lower()
                        
                if len(letra_jugada)>1:
                                print(f"Has introducido \" {letra_jugada.upper()} \", solo puedes jugar con UNA letra")
                
                elif letra_jugada.isalpha() != True:
                        print(f"Has introducido \" {letra_jugada} \", solo puedes jugar con LETRAS")

                
                elif letra_jugada in palabra_adivinar:
                        for i in range(len(palabra_adivinar)):
                                if palabra_adivinar[i] == letra_jugada:
                                        palabra_incognito[i] = letra_jugada  
                                        secreta = "".join(palabra_incognito)                                

                else:  
                        letras_falladas.append(letra_jugada) 
                        vidas-=1


                
                print(secreta)
                print(f"❌Letras falladas: {letras_falladas}")
                print("\n___________________________________________________")
                print(muñeco[vidas])
                

        if secreta == palabra_adivinar:
                print(f"ENHORABUENA!!! HAS GANADO 🏆, la palabra era {palabra_adivinar}")

        else:
                print(f"Oh! No has acertado, la palabra era {palabra_adivinar}")
    


```
