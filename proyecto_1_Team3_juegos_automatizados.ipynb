```python

class Juegos:

    def imprimir_tablero(self,tablero):
        print("******************")
        print(tablero[0][0],"|",tablero[0][1],"|",tablero[0][2])
        print("__________")
        print(tablero[1][0],"|",tablero[1][1],"|",tablero[1][2])
        print("__________")
        print(tablero[2][0],"|",tablero[2][1],"|",tablero[2][2])


        #Creamos la funcion imprimir tablero para llamarla en el juego de 3 en raya, y de esta forma automatizarlo, necesita de las variables del juego para poder crearlo.
        #Nos gustaria que fuese "privada" y solo poder acceder a ella dentro del juego.

    def tres_en_raya(self):
        jugar = True
        tablero = [[' ' for _ in range(3)] for _ in range(3)]
        jugada= 0
        ficha_jugador1= "X"
        ficha_jugador2= "O"

        #Creamos las variables necesarias para el juego (condicion de ejecucion del codigo, recuento de jugadas, tablero y jugadores)


        #Establecemos la condicion de juego y asignamos el turno de jugador, el jugador uno empezara la partida y jugara los turnos pares, y el jugador 2 los impares.
        #Limitamos los inputs del jugador para que no haya errores (letras, decimales, caracteres, SOLO numeros del 0 al 2)

        while jugar:
            if jugada %2 == 0:
                ficha_tablero = ficha_jugador1
                
            else:
                ficha_tablero = ficha_jugador2

            try:
                fila = int(input(f"Introduce FILA {ficha_tablero}"))
                columna = int(input(f"Introduce COLUMNA {ficha_tablero}"))

            except:
                print("Dato erroneo, solo puede introducir NUMEROS del 1 al 2")
                continue

            
            if fila <0 or fila >2:
                print("Error, valor de FILA incorrecto (selecciona un numero 0, 1 o 2)")
                continue
            
            if columna <0 or columna >2:
                print("Error, valor de COLUMNA incorrecto (selecciona un numero 0, 1 o 2)")
                continue

            if tablero[fila][columna] != " ":
                print("Casilla ocupada")
                self.imprimir_tablero(tablero)
                continue # Que no ponga ficha si la casilla ya esta ocupada
                

            tablero[fila][columna]= ficha_tablero

            #establecemos las posibilidades que hay de ganar el juego. 
            #Por cada una de ellas, validamos la fecha y revisamos si continua la partida (Comenzamos de nuevo el bucle) o hay un ganador
               
            if tablero[0][0] == tablero[0][1] and tablero[0][0] == tablero[0][2] and tablero[0][0] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[1][0] == tablero[1][1] and tablero[1][0] == tablero[1][2] and tablero[1][0] != " ":
                    if tablero[1][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[2][0] == tablero[2][1] and tablero[2][0] == tablero[2][2] and tablero[2][0] != " ":
                    if tablero[2][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False
                    
            elif tablero[0][0] == tablero[1][0] and tablero[0][0] == tablero[2][0] and tablero[0][0] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[0][1] == tablero[1][1] and tablero[0][1] == tablero[2][1] and tablero[0][1] != " ":
                    if tablero[0][0] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[0][2] == tablero[1][2] and tablero[0][2] == tablero[2][2] and tablero[0][2] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[0][0] == tablero[1][1] and tablero[0][0] == tablero[2][2] and tablero[0][0] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False

            elif tablero[0][2] == tablero[1][1] and tablero[0][2] == tablero[2][0] and tablero[0][2] != " ":
                    if tablero[0][2] == ficha_jugador1:
                        print("Enhorabuena ha ganado el JUGADOR 1 üéâ")
                    else:
                        print("Enhorabuena ha ganado el JUGADOR 2 üèÜ")
                    jugar = False
            
            elif jugada == 8:        
                print("Vaya, parece que los jugadores han quedado en tablas!üò∞")
                jugar = False
            
            #llamamos a la funcion para que nos muestre el tablero y sume en el contador de jugadas para que pase el turno al siguiente jugador
                
            self.imprimir_tablero(tablero)
            jugada += 1



    def piedra_papel_tijera(self):

        #Para jugar a piedra/papel tenenos que importar la libreria random que esta en la celda superior de la clase
        #Establecemos las variables de juego (Opciones, jugadores, contador...)

        lista_opciones = ["Piedra","Papel","Tijera"]

        partidas_maximas = 3
        contador_jugador1 = 0
        contador_jugador2 = 0
        partida = 0

        #Establecemos la condicion de juego
        #Jugaremos contra la maquina por lo que el "jugador1" se selecciona a traves de las funciones de la libreria random, mientras que el "jugador2" lo realiza a travesde un input 

        while partida < partidas_maximas:
            jugador2 = input("Elige para jugar Piedra, Papel, Tijera").lower()
            jugador1 = random.choice(lista_opciones).lower()

            if jugador2.title() not in lista_opciones:
                print("‚ùåOpcion no valida, prueba de nuevo con PIEDRA/PAPEL/TIJERA ")
                continue
                         
        #Establecemos las posibilidades de juego  para validar con los argumentos introducidos por los jugadores
            
            if jugador1 == jugador2:
                print("Oh, habeis quedado EMPATE!")
                print(f"Ambos elegisteis {jugador2}")
                
            
            elif (jugador1 == "piedra" and jugador2 == "tijera") or (jugador1 == "papel" and jugador2 == "piedra") or (jugador1 == "tijera" and jugador2 == "papel"):
                
                contador_jugador1 +=1 
                print("¬°Partida ganada por el ordenador! üíª")
                print(f"El ordenador eligio {jugador1} contra tu eleccion {jugador2}")
                

            else:
                contador_jugador2 +=1
                print("¬°Partida ganada por el jugador!ü•á")
                print(f"El ordenador eligio {jugador1} contra tu eleccion {jugador2}")
                
            
            partida = contador_jugador1 + contador_jugador2


         #Dise√±amos un mensaje para anunciar el ganador al terminar la partida   

            print(f"Partidas jugadas: {partida}")
            print("______________________________")
        print("**********************************")
        if contador_jugador1 > contador_jugador2:
            print("Ha ganador EL ORDENADOR!üò≠")
        else:
            print("ENHORABUENA!!! Has ganado üèÜ")

    
    def ahorcado(self):
        
        #Estructuramos el juego para que sea entre 2 jugadores, (Jugador uno establece una palabra, jugador 2 encargado de adivinarla)

        #Establecemos las variables necesarias para la partida (vidas/oportunidades, mu√±eco del ahorado, las letras falladas...)

        palabra_adivinar = input("elige una palabra").lower() #jugador 1
        letras_falladas = []
        vidas= 6
        palabra_incognito = ["_"]*len(palabra_adivinar)
        secreta = ""
        mu√±eco= ["------\n|  |\n|  O\n| /|\\\n| / \\\n|------\n",
                "------\n|  |\n|  O\n| /|\\\n| / \n|------\n",
                "------\n|  |\n|  O\n| /|\\\n| \n|------\n",
                "------\n|  |\n|  O\n| /|\n| \n|------\n",
                "------\n|  |\n|  O\n| /\n| \n|------\n",
                "------\n|  |\n|  O\n| \n| \n|------\n",
                "------\n|  |\n|  \n| \n| \n|------\n"]

        #Condicion de juego para que entre en el bucle

        while vidas > 0 and secreta!= palabra_adivinar:

        #El jugador 2 introducira una letra que procedera a validar
                letra_jugada = input("Introducce una letra").lower()
        
        #Primero validamos que sea una unica letra y que no sean numeros, caracteres...
                        
                if len(letra_jugada)>1:
                                print(f"Has introducido \" {letra_jugada.upper()} \", solo puedes jugar con UNA letra")
                
                elif letra_jugada.isalpha() != True:
                        print(f"Has introducido \" {letra_jugada} \", solo puedes jugar con LETRAS")

        #Comprobamos que la letra esta en la palabra del jugador 1 
        #Recorremos todas las posiciones de la palabra adivinar
        #Verificamos en cada posicion que la letra que hemos jugado es igual a la palabra 
        #Si la letra coincide con la posicion [i] la revelamos en la palabra incognito
        #Actualizamos la variable "secreta" de lista a cadena de string
                
                elif letra_jugada in palabra_adivinar:
                        for i in range(len(palabra_adivinar)):
                                if palabra_adivinar[i] == letra_jugada:
                                        palabra_incognito[i] = letra_jugada  
                                        secreta = "".join(palabra_incognito)                                

                else:  
                        letras_falladas.append(letra_jugada) 
                        vidas-=1


                
                print(secreta)
                print(f"‚ùåLetras falladas: {letras_falladas}")
                print("\n___________________________________________________")
                print(mu√±eco[vidas])
                

        if secreta == palabra_adivinar:
                print(f"ENHORABUENA!!! HAS GANADO üèÜ, la palabra era {palabra_adivinar}")

        else:
                print(f"Oh! No has acertado, la palabra era {palabra_adivinar}")
    


```
